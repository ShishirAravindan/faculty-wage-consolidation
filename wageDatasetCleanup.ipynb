```python
# I want to read the excel file in this directory called a.xlsx and show the header
def read_excel_file(file_name):
    """
    Reads an Excel file (.xlsx) from the current directory,
    loads it into a pandas DataFrame, and prints the header.

    Parameters:
        file_name (str): Name of the Excel file to be loaded.
    """
    df = pd.read_excel(file_name)
    display(df.head(5))
    return df
```

```python
historical_df = read_excel_file(HISTORICAL_DATASET)
historical_df.shape
```

```python
cleaned_df = read_excel_file(CLEANED_DATASET)
cleaned_df.shape
```

```python
# Rename the column in the dataframes to Rank
historical_df = historical_df.rename(columns={'Class': 'Rank'})
cleaned_df = cleaned_df.rename(columns={'Class': 'Rank'})
```

```python
updateCount = 0
for index, row in cleaned_df.iterrows():
    name, dept = row['Name'], row['Department']
    mask = historical_df['Name'] == name
    updateMask = mask & historical_df['Department'].isna()
    historical_df.loc[updateMask, 'Department'] = dept
    updateCount += updateMask.sum()
print(updateCount)
```

```python
import re

def replace_department(dept):
    if isinstance(dept, str):
        if re.search(r'\bFinance\b', dept, re.IGNORECASE):
            return 'FIN'
        elif re.search(r'\bBusiness Administration\b', dept, re.IGNORECASE):
            return 'BUSI'
    return dept

historical_df['Department'] = historical_df['Department'].apply(replace_department)
```

```python
# Merging tippie data to dataset
tippie_df = pd.read_excel('tippie.xlsx')

updateCount = 0
for index, row in tippie_df.iterrows():
    name, dept = row['Name'], row['Faculty']
    print(f"{index} | {name}")
    mask = historical_df['Name'] == name
    updateMask = mask & historical_df['Department'].isna()
    historical_df.loc[updateMask, 'Department'] = dept
    updateCount += updateMask.sum()
print(updateCount)
```
